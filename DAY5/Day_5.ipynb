```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("nuclear_explosions.csv")
```

```python
# Print the columns of the data
print("Columns of the data:")
print(data.columns)
```

```python
data.rename(columns={
    'WEAPON SOURCE COUNTRY': 'Source_Country',
    'WEAPON DEPLOYMENT LOCATION': 'Deployment_Location',
    'Data.Source': 'Source',
    'Location.Cordinates.Latitude': 'Latitude',
    'Location.Cordinates.Longitude': 'Longitude',
    'Data.Magnitude.Body': 'Magnitude_Body',
    'Data.Magnitude.Surface': 'Magnitude_Surface',
    'Location.Cordinates.Depth': 'Depth',
    'Data.Yeild.Lower': 'Yield_Lower',
    'Data.Yeild.Upper': 'Yield_Upper',
    'Data.Purpose': 'Data_Purpose',
    'Data.Name': 'Data_Name',
    'Data.Type': 'Data_Type',
    'Date.Day': 'Day',
    'Date.Month': 'Month',
    'Date.Year': 'Year'
}, inplace=True)
```

```python
# Check for missing values and treat them
print("Missing values:")
print(data.isna().sum())

# Replace string "NAN" with actual NaN values
data.replace("NAN", pd.NA, inplace=True)

# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Unique values in categorical columns
categorical_columns = ['Source_Country', 'Deployment_Location', 'Source', 'Data_Purpose', 'Data_Type']
for col in categorical_columns:
    unique_values = data[col].unique()
    print(f"Unique values in {col}:")
    print(unique_values)
```

```python
# Check data types of each column
print("Data types of each column:")
print(data.dtypes)
```

```python
# Country-wise count of nuclear explosions
country_explosion_count = data['Source_Country'].value_counts()
country_explosion_count.plot(kind="bar", figsize=(10, 6))
plt.title("Source Country-wise Count of Nuclear Explosions")
plt.xlabel("Source Country")
plt.ylabel("Count")
plt.show()
```

```python
# Deployment location count
deployment_location_count = data['Deployment_Location'].value_counts()
deployment_location_count.plot(kind="bar", figsize=(10, 6))
plt.title("Deployment Location Count")
plt.xlabel("Deployment Location")
plt.ylabel("Count")
plt.show()
```

```python
# Year-wise count
year_explosion_count = data['Year'].value_counts().sort_index()
year_explosion_count.plot(kind="bar", figsize=(10, 6))
plt.title("Year-wise Count of Nuclear Explosions")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# Country-wise average explosion yield
country_avg_yield = data.groupby('Source_Country')['Yield_Lower'].mean()
country_avg_yield.plot(kind="bar", figsize=(10, 6))
plt.title("Source Country-wise Average Explosion Yield")
plt.xlabel("Source Country")
plt.ylabel("Average Yield")
plt.show()
```

```python
# Data Purpose Distribution
data_purpose_distribution = data['Data_Purpose'].value_counts()
data_purpose_distribution.plot(kind="bar", figsize=(10, 6))
plt.title("Data Purpose Distribution")
plt.xlabel("Data Purpose")
plt.ylabel("Frequency")
plt.show()
```

```python
# Correlation between yield and depth of nuclear test
plt.figure(figsize=(10, 6))
plt.scatter(data['Yield_Lower'], data['Depth'], alpha=0.5)
plt.title("Correlation between Yield and Depth of Nuclear Tests")
plt.xlabel("Yield (Lower Bound)")
plt.ylabel("Depth")
plt.show()
```

```python

# Box plot of all numeric variables
data_numeric = data.select_dtypes(include=['float64', 'int64'])
data_numeric.boxplot(figsize=(12, 8))
plt.title("Box Plot of Numeric Variables")
plt.xticks(rotation=45)
plt.show()
```
